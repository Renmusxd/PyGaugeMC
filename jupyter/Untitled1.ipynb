```python
def dims_from_p(pi):
    return [
        (i,j) for i in range(4) for j in range(i+1,4)
    ][pi]

def f(index, t=4, x=4, y=4, z=4, p=6):
    pi = index // (t*x*y*z // 4)
    subindex = index % (t*x*y*z // 4)
    
    bounds = [t,x,y,z]
    mu, nu = dims_from_p(pi)
    bounds[mu] //= 2
    bounds[nu] //= 2
    
    ti = subindex // (bounds[1] * bounds[2] * bounds[3])
    subindex = subindex % (bounds[1] * bounds[2] * bounds[3])
    xi = subindex // (bounds[2] * bounds[3])
    subindex = subindex % (bounds[2] * bounds[3])
    yi = subindex // (bounds[3])
    subindex = subindex % (bounds[3])
    zi = subindex
    coords = [ti,xi,yi,zi]
    coords[mu] *= 2
    coords[nu] *= 2
    return coords, pi
```

```python
for i in range(4*4*4*4*6 // 4):
    print(f(i))
```
